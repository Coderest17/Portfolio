```python
def move_chance():
    global chance_cards
    temp = chance_cards[0]
    chance_cards = chance_cards[1:]
    chance_cards.append(temp)
```

```python
def move_chest():
    global chest_cards
    temp = chest_cards[0]
    chest_cards = chest_cards[1:]
    chest_cards.append(temp)
```

```python
def move(roll):
    global current
    global board
    global streets
    global chest_cards
    global chance_cards
    
    current += roll
    if current >= 40:
        current -= 40
    streets[board[current]] += 1
    
    if board[current] == "Go To Jail":
        current = board.index("Jail")
        streets["Jail"] += 1
    
    elif board[current].split()[0] == "Chance":
        temp = chance_cards[0]
        move_chance()
        if temp == "GO":
            current = board.index("GO")
            streets["GO"] += 1
        elif temp == "Jail":
            current = board.index("Jail")
            streets["Jail"] += 1
        elif temp == "Illinois Avenue":
            current = board.index("Illinois Avenue")
            streets["Illinois Avenue"] += 1
        elif temp == "St. Charles Place":
            current = board.index("St. Charles Place")
            streets["St. Charles Place"] += 1
        elif temp == "Reading Railroad":
            current = board.index("Reading Railroad")
            streets["Reading Railroad"] += 1
        elif temp == "Boardwalk":
            current = board.index("Boardwalk")
            streets["Boardwalk"] += 1
        elif temp == "NEAREST UTILITY":
            while board[current].split()[-1] != "Utility":
                current += 1
                if current == 40:
                    current = 0
            streets[board[current]] += 1
        elif temp == "NEAREST RAILROAD":
            while board[current].split()[-1] != "Railroad":
                current += 1
                if current == 40:
                    current = 0
            streets[board[current]] += 1
        elif temp == "BACK 3":
            current -= 3
            streets[board[current]] += 1
        
    elif board[current].split()[0] == "Community":
        temp = chest_cards[0]
        move_chest()
        if temp == "GO":
            current = board.index("GO")
            streets["GO"] += 1
        elif temp == "Jail":
            current = board.index("Jail")
            streets["Jail"] += 1
```

```python
def check_history(roll):
    global history
    first = [0,0,0]
    second = [0,0,0]
    result = [0,0,0]
    if len(history) <= 2:
        history.append(roll)
    else:
        del history[0]
        history.append(roll)
        
    if len(history) == 3:
        for i in range(3):
            first[i] = history[i][0]
            second[i] = history[i][1]
            
    for q in range(3):
        if first[q] == second[q] and first[q] != 0:
            result[q] = 1
            
    if sum(result) == 3:
        print(history)
        history = []
        return True
    return False
```

```python
def move_new(roll):
    #This move_new function also takes doubles into account....
    global current
    global board
    global streets
    global chest_cards
    global chance_cards
    global history
    
    c = check_history(roll)
    if c:
        current = board.index("Jail")
        streets["Jail"] += 1
    else:
        a = roll[0]
        b = roll[1]
        current += a
        current += b
        if current >= 40:
            current -= 40
        streets[board[current]] += 1

        if board[current] == "Go To Jail":
            current = board.index("Jail")
            streets["Jail"] += 1

        elif board[current].split()[0] == "Chance":
            temp = chance_cards[0]
            move_chance()
            if temp == "GO":
                current = board.index("GO")
                streets["GO"] += 1
            elif temp == "Jail":
                current = board.index("Jail")
                streets["Jail"] += 1
            elif temp == "Illinois Avenue":
                current = board.index("Illinois Avenue")
                streets["Illinois Avenue"] += 1
            elif temp == "St. Charles Place":
                current = board.index("St. Charles Place")
                streets["St. Charles Place"] += 1
            elif temp == "Reading Railroad":
                current = board.index("Reading Railroad")
                streets["Reading Railroad"] += 1
            elif temp == "Boardwalk":
                current = board.index("Boardwalk")
                streets["Boardwalk"] += 1
            elif temp == "NEAREST UTILITY":
                while board[current].split()[-1] != "Utility":
                    current += 1
                    if current == 40:
                        current = 0
                streets[board[current]] += 1
            elif temp == "NEAREST RAILROAD":
                while board[current].split()[-1] != "Railroad":
                    current += 1
                    if current == 40:
                        current = 0
                streets[board[current]] += 1
            elif temp == "BACK 3":
                current -= 3
                streets[board[current]] += 1

        elif board[current].split()[0] == "Community":
            temp = chest_cards[0]
            move_chest()
            if temp == "GO":
                current = board.index("GO")
                streets["GO"] += 1
            elif temp == "Jail":
                current = board.index("Jail")
                streets["Jail"] += 1
```
